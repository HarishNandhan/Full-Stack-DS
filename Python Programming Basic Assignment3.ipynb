```python
# Q1.Write a Python Program to Check if a Number is Positive, Negative or Zero?

for i in range(0,3):
    a=int(input("Enter a number: "))

    if a>0:
        print("Positive number")
    elif a<0:
        print("Negative number")
    else:
        print("Zero")
```

```python
# Q2.Write a Python Program to Check if a Number is Odd or Even?
for i in range(0,2):
    a=int(input("Enter a number: "))

    if a%2==0:
        print("Even number")
    else:
        print("Odd number")
```

```python
# Q3.Write a Python Program to Check Leap Year?

a=int(input("Enter a year: "))
if a%400==0 and a%100==0:
    print("Leap year")
elif a%4==0 and a%100!=0:
    print("Leap year")
else:
    print("Not a Leap year")
```

```python
# Q4.Write a Python Program to Check Prime Number?

a=int(input("Enter a number: "))
if a==1:
    print("Neither a prime number nor a composite number")
elif a==2 or a==3 or a==5 or a==7:
    print("Prime number")
elif a%2!=0 and a%3!=0 and a%5!=0 and a%7!=0:
    print("Prime number")
else:
    print("Not a prime number")
```
